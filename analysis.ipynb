```python
import sympy as sp
from sympy.plotting import plot3d, plot_implicit, PlotGrid
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu_s = 1.15 # coefficient of friction between rubber and rubber TODO: check this experimentally
w = 5 # max weight in lbs
g = 32.2 # gravitational acceleration in ft/s^2
```

```python
%matplotlib widget
T = sp.symbols('T') # torque in ft-lbs
omega = sp.symbols('omega') # angular velocity in rpm

N = w # normal force in lbs
F = N * mu_s # max frictional force in lbs

r = T / F # max radius of the wheel
mph = omega * r * 60 / 5280 # max speed in mph

# plot the max radius of the wheel as a function of the motors torque rating
sp.plot(r*12, (T, 0, 10), xlabel='Motor Torque (ft-lbs)', ylabel='Max Wheel Radius (inches)', title='Standard wheels')

# plot the max speed of the robot as a function of the motors torque rating and angular velocity in 3D
plot3d(mph, (T, 0, 10), (omega, 20, 100), xlabel='Motor Torque (ft-lbs)', ylabel='Motor Angular Velocity (rpm)', zlabel='Max Vehicle Speed (mph)', title='Standard wheels')

r1 = 3*T / F # max radius of the wheel
mph1 = omega * r1 * 60 / 5280 # max speed in mph
# plot the max radius of the wheel as a function of the motors torque rating
sp.plot(r1*12, (T, 0, 10), xlabel='Motor Torque (ft-lbs)', ylabel='Max Wheel Radius (inches)', title='Gear ratio of 1:3')

# plot the max speed of the robot as a function of the motors torque rating and angular velocity in 3D
plot3d(mph1, (T, 0, 10), (omega, 20, 100), xlabel='Motor Torque (ft-lbs)', ylabel='Motor Angular Velocity (rpm)', zlabel='Max Vehicle Speed (mph)', title='Gear ratio of 1:3')
```

```python
_T = 40 # kgf*cm
_omega = 30 # rpm

# convert to ft*lbs
_T = _T * 0.07233013851
_maxR = r.subs(T, _T)
_maxSpeed = mph.subs(T, _T).subs(omega, _omega)

print("------------NO GEAR RATIO------------")
print(f'Max required wheel diameter: {2*_maxR.evalf()} inches')
print(f'Max speed: {_maxSpeed.evalf()} mph')
print(f'Max speed: {_maxSpeed.evalf() * 44.704} cm/s')
```
